```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/kaggle/input/gym-exercise-data/megaGymDataset.csv")
```

# # one_hot_encoder

```python
column=df[["Type","BodyPart","Equipment"]]
one_hot=OneHotEncoder(sparse=False)
one=one_hot.fit_transform(column)
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
```

# # vectorizer

```python
tfid=TfidfVectorizer()
tf=tfid.fit_transform(df["Title"])
tf=tf.toarray()
tf=pd.DataFrame(tf,columns=tfid.get_feature_names_out())
```

```python
concaty=pd.concat([one,tf],axis=1)
Y=df["Level"]
```

```python
df
```

# # model training

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,Y,test_size=0.2,random_state=42)
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
```

```python
random_classif.score(X_test,Y_test)
```
