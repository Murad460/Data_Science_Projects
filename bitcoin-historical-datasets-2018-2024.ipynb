```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df_15m=pd.read_csv("/kaggle/input/bitcoin-historical-datasets-2018-2024/btc_15m_data_2018_to_2024-09-09.csv")
df_1d=pd.read_csv("/kaggle/input/bitcoin-historical-datasets-2018-2024/btc_1d_data_2018_to_2024-09-06.csv")
df_1h=pd.read_csv("/kaggle/input/bitcoin-historical-datasets-2018-2024/btc_1h_data_2018_to_2024-09-06.csv")
df_4h=pd.read_csv("/kaggle/input/bitcoin-historical-datasets-2018-2024/btc_4h_data_2018_to_2024-09-06.csv")
```

```python
def col_datetime(df,col):
    df[col]=pd.to_datetime(df[col])
    df["Year"]=df[col].dt.year
    df["Month"]=df[col].dt.month
    df["Day"]=df[col].dt.day
    df["Hour"]=df[col].dt.hour
    df["Minute"]=df[col].dt.minute
    df["Second"]=df[col].dt.second
    return df
# Y=df_15m["Taker buy base asset volume"]
```

```python
df2=col_datetime(df_15m,"Open time")
df3=col_datetime(df_15m,"Open time")
df4=col_datetime(df_15m,"Open time")
df5=col_datetime(df_15m,"Open time")
```

```python
df32=col_datetime(df2,"Close time")
df32=df3.drop(columns=["Open time","Close time","Taker buy base asset volume"])

df33=col_datetime(df3,"Close time")
df33=df3.drop(columns=["Open time","Close time","Taker buy base asset volume"])

df34=col_datetime(df4,"Close time")
df34=df3.drop(columns=["Open time","Close time","Taker buy base asset volume"])

df35=col_datetime(df5,"Close time")
df35=df3.drop(columns=["Open time","Close time","Taker buy base asset volume"])
```

```python
def split_data(X,Y):
    
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
    return X_train,X_test,Y_train,Y_test
```

```python
X_train1,X_test1,Y_train1,Y_test1=split_data(df32,df_15m["Taker buy base asset volume"])
X_train2,X_test2,Y_train2,Y_test2=split_data(df33,df_15m["Taker buy base asset volume"])
X_train3,X_test3,Y_train3,Y_test3=split_data(df34,df_15m["Taker buy base asset volume"])
X_train4,X_test4,Y_train4,Y_test4=split_data(df35,df_15m["Taker buy base asset volume"])
```

```python
random_reg=RandomForestRegressor()
random_reg.fit(X_train1,Y_train1)
random_reg.score(X_test1,Y_test1)
```
