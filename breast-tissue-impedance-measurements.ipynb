# # Machine learnning

```python
lab_encod=LabelEncoder()
df["Class_encod"]=lab_encod.fit_transform(df["Class"])
new_df=df.drop(columns=["Class"])
X2=new_df.drop(columns=["Class_encod"])
Y2=new_df[["Class_encod"]]
```

```python
X_train2,X_test2,Y_train2,Y_test2=train_test_split(X2,Y2,test_size=0.2,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Input,Flatten,Dropout,LSTM,SimpleRNN,GRU
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
                monitor='val_loss',
                patience=5,
                verbose=1)
```

```python
model=Sequential([
    Input(shape=(9,)),
    Dense(32,activation="relu"),
    Dense(128,activation="relu"),
    Dropout(0.5),
    Dense(64,activation="relu"),
    Dense(34,activation="relu"),
    Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train2,Y_train2,epochs=100,batch_size=32,validation_data=(X_test2,Y_test2),validation_split=0.2,callbacks=[early_stopping])
```
