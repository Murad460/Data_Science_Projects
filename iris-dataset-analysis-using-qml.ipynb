```python
import tensorflow as tf
import tensorflow_quantum as tfq
import cirq
import sympy
import numpy as np

# Create a simple quantum circuit with a parameterized gate
qubit = cirq.GridQubit(0, 0)
theta = sympy.Symbol('theta')
circuit = cirq.Circuit(cirq.rx(theta)(qubit))

# Prepare a batch of circuits
num_samples = 5
circuit_batch = tfq.convert_to_tensor([circuit] * num_samples)

# Define the readout operator
readout_op = cirq.Z(qubit)
readout_op_batch = tfq.convert_to_tensor([[readout_op]] * num_samples)

# Create a TFQ Expectation layer
expectation_layer = tfq.layers.Expectation()

# Prepare input data: symbol values for theta
theta_values = np.array([[0.0], [np.pi / 2], [np.pi], [3 * np.pi / 2], [2 * np.pi]])
theta_tensor = tf.convert_to_tensor(theta_values, dtype=tf.float32)

# Calculate the expectation values
expectation_values = expectation_layer(
    circuit_batch, symbol_names=['theta'], symbol_values=theta_tensor, operators=readout_op_batch)

# Print the results
print("Theta values:\n", theta_values)
print("Expectation values:\n", expectation_values.numpy())

# Check if TFQ is working properly by comparing expected results
expected_values = np.cos(theta_values)
print("Expected values:\n", expected_values)
print("Difference:\n", np.abs(expectation_values.numpy() - expected_values))
```
