```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords 
nltk.download("punkt")
nltk.download("stopwords")
```

```python
df=pd.read_csv("/kaggle/input/imdb-top-1000-tv-series/data.csv")
```

```python
df.nunique()
```

```python
df
```

```python
df["genres"]=df['genres'].str.split(',')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(df['genres'])
```

```python
genre_encoded
```

```python
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)
genre_df
```

```python
def text_preprocess(text):
    data=text.lower()
    data_re=re.sub(r"\W+"," ",data)
    data_word_tokenize=word_tokenize(data_re)
    token=[word for word in data_word_tokenize if word not in stopwords.words("english")]
    return " ".join(token)
```

```python
df["title"]=df["title"].apply(text_preprocess)
```

```python
df.sample()
```

```python
def text_vectorizer(text):
    vectorizer=CountVectorizer()
    vect=vectorizer.fit_transform(text)
    return pd.DataFrame(vect.toarray(),columns=vectorizer.get_feature_names_out())
```

```python
dd=text_vectorizer(df["title"])
```

```python
dd
```

```python
concaty=pd.concat([df,dd,genre_df],axis=1).drop(columns=["title","genres","id"])
```

```python
concaty
```

```python
df["averageRating"].nunique()
```

```python
X=concaty.drop(columns=["averageRating"])
Y=concaty["averageRating"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
Y_train
```

# # Machine Learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
models=[RandomForestRegressor(),DecisionTreeRegressor(),KNeighborsRegressor()]
for mod in models:
    model=mod
    model.fit(X_train,Y_train)
    print(f"The accuracy of {mod} is ",model.score(X_test,Y_test))
    
```
