```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("/kaggle/input/parkinsons-disease-data-set/parkinsons.data",delimiter=',')
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=["name","status"])
Y=df[["status"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
y_pred=random_classif.predict(X_test)
y_pred
```

```python
sns.histplot(Y_test)
sns.histplot(y_pred,color="red")
```
