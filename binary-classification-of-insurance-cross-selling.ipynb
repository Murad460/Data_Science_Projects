# # Deep learning

```python
from keras.layers import Dense,Input,Dropout,BatchNormalization
from keras.models import Model
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
X_train.shape
```

```python
X_train2=X_train.iloc[0:2000000]
Y_train2=Y_train.iloc[0:2000000]
Y_train.unique()
X_train3=X_train.iloc[2000000:4000000]
Y_train3=Y_train.iloc[2000000:4000000]

X_train4=X_train.iloc[4000000:6000000]
Y_train4=Y_train.iloc[4000000:6000000]

X_train5=X_train.iloc[6000000:9000000]
Y_train5=Y_train.iloc[6000000:9000000]
```

# # # M2

```python
inputs=Input(shape=(14,))
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01),name="dense_64")(d)
b=BatchNormalization()(d)
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
d=Dropout(0.5)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01),name="dense_64")(d)
outputs=Dense(1,activation="sigmoid")(b)
model=Model(inputs,outputs)
model.compile(optimizer="adam",loss='binary_crossentropy',metrics=["accuracy"])

for layer in model.layers:
    layer.trainable=False
dense_output=layer.output

d2=Dense(32,activation="relu")(dense_output)
d2=Dense(128,activation="relu")(d2)
d2=Dense(256,activation="relu")(d2)
outputs2=Dense(1,activation="sigmoid")(d2)
new_model=Model(inputs=model.input,outputs=outputs2)
new_model.compile(optimizer="adam",loss='binary_crossentropy',metrics=["accuracy"])
new_model.summary()
```

```python
# ,(X_train4,Y_train4),(X_train5,Y_train5)
```

```python
for mod,X_tra,Y_tra in [(model,X_train2,Y_train2),(new_model,X_train3,Y_train3)]:
    mod.fit(X_tra,Y_tra,epochs=50,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

# # # M3

```python
inputs=Input(shape=(14,))
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
b=BatchNormalization()(d)
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(b)

outputs=Dense(1,activation="sigmoid")(d)
modeld3=Model(inputs,outputs)
modeld3.compile(optimizer="adam",loss='binary_crossentropy',metrics=["accuracy"])
```

```python
for X_tra3,Y_tra3 in [(X_train2,Y_train2),(X_train3,Y_train3),(X_train4,Y_train4),(X_train5,Y_train5)]:
    modeld3.fit(X_tra3,Y_tra3,epochs=50,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```
