```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/kaggle/input/financial-risk-for-loan-approval/Loan.csv")
```

```python
df_columns_obj=df.select_dtypes(include=["object","category"]).columns
new_df=pd.get_dummies(df,columns=df_columns_obj,drop_first=False).astype(int)

```

```python
df
```

```python
X=new_df.drop(columns=["RiskScore"])
Y=new_df["RiskScore"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_reg=RandomForestRegressor()
random_reg.fit(X_train,Y_train)
random_reg.score(X_test,Y_test)
```

```python
log_reg=LogisticRegression()
log_reg.fit(X_train,Y_train)
log_reg.score(X_test,Y_test)
```
