```python
from keras.layers import Input,Dense,Dropout,LSTM,SimpleRNN,GRU,TimeDistributed
from keras.models import Model
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
X_train_reshaped=X_train.values.reshape(X_train.shape[0],1,X_train.shape[1])
X_test_reshaped=X_test.values.reshape(X_test.shape[0],1,X_test.shape[1])
X_train_reshaped.shape
```

```python
X_train.shape
```

```python
inputs=Input(shape=(21,))
D=Dense(32,activation="relu")(inputs)
D=Dense(64,activation="relu")(D)
D=Dense(128,activation="relu")(D)
do=Dropout(0.5)(D)
D=Dense(32,activation="relu")(do)
D=Dense(64,activation="relu")(D)
D=Dense(128,activation="relu")(D)
do=Dropout(0.5)(D)
D=Dense(32,activation="relu")(do)
outputs=Dense(1,activation="sigmoid")(D)
model=Model(inputs,outputs)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=50,batch_size=32,validation_data=(X_test,Y_test),callbacks=[early_stopping])
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
```

```python
from keras.regularizers import l2
from keras.layers import Input, SimpleRNN, LSTM, GRU, Dense, Dropout, BatchNormalization, TimeDistributed
from keras.models import Model
from keras.initializers import HeNormal


inputs = Input(shape=(1, 21))


S = SimpleRNN(32, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
S = BatchNormalization()(S)
S = SimpleRNN(64, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(S)
S = BatchNormalization()(S)
S = SimpleRNN(64, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(S)
S = BatchNormalization()(S)


T = TimeDistributed(Dense(32, activation='relu', kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)))(S)
T = BatchNormalization()(T)
T = TimeDistributed(Dense(64, activation='relu', kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)))(T)
T = BatchNormalization()(T)


do = Dropout(0.5)(T)


L = LSTM(64, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(do)
L = BatchNormalization()(L)
L = LSTM(128, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L = BatchNormalization()(L)
L = LSTM(128, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L = BatchNormalization()(L)


T = TimeDistributed(Dense(128, activation='relu', kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)))(L)
T = BatchNormalization()(T)
T = TimeDistributed(Dense(64, activation='relu', kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)))(T)
T = BatchNormalization()(T)


do = Dropout(0.5)(T)


G = GRU(64, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(do)
G = BatchNormalization()(G)
G = GRU(128, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(G)
G = BatchNormalization()(G)
G = GRU(128, return_sequences=True, kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(G)
G = BatchNormalization()(G)


outputs = TimeDistributed(Dense(1, activation='sigmoid', kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)))(G)


model2S6T4L6T4G6kr = Model(inputs, outputs)
model2S6T4L6T4G6kr.compile(optimizer='adam', loss='binary_crossentropy', metrics=["accuracy"])


model2S6T4L6T4G6kr.summary()
```
