```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/kaggle/input/full-filled-brain-stroke-dataset/full_data.csv")
# df2=pd.read_csv("/kaggle/input/full-filled-brain-stroke-dataset/full_filled_stroke_data (1).csv")
```

```python
df
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["gender","ever_married","work_type","Residence_type","smoking_status"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
one
```

```python
concaty=pd.concat([df,one],axis=1).drop(columns=["gender","ever_married","work_type","Residence_type","smoking_status","stroke"])
```

```python
concaty
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,df["stroke"],test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```
