```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
df=pd.read_csv("/kaggle/input/wisconsin-breast-cancer-cytology-features/wisconsin_breast_cancer.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(how="any",axis=0)
df
```

```python
sns.heatmap(df.corr())
```

```python
X=df.drop(columns=["id","class"])
Y=df["class"]
```

```python
Y.unique()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
for i in [RandomForestClassifier(),GradientBoostingClassifier(),AdaBoostClassifier(),BaggingClassifier(),SVC(),LogisticRegression()]:
    random_classif=i
    random_classif.fit(X_train,Y_train)
    print(f"the accurac of{i}:",random_classif.score(X_test,Y_test))
```
