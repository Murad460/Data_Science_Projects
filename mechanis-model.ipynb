```python
label_encoder=LabelEncoder()
df["sig_id_new"]=label_encoder.fit_transform(df["sig_id"])
df["drug_id_new"]=label_encoder.fit_transform(df["drug_id"])
df2["cp_type_new"]=label_encoder.fit_transform(df2["cp_type"])
df2["cp_dose_new"]=label_encoder.fit_transform(df2["cp_dose"])
df_new=df.drop(columns=["sig_id","drug_id"])
df_new
```

```python
df2_new=df2.drop(columns=["sig_id","cp_type","cp_dose"])
```

```python
X=pd.concat([df_new,df2_new],axis=1)
Y1=pd.concat([df3,df4],axis=1)
Y=Y1.drop(columns=["sig_id"])
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
reg=RandomForestClassifier()
reg.fit(X_train,Y_train)
```

```python
grad_reg=GradientBoostingClassifier()
grad_reg.fit(X_train,Y_train)
```

```python
ada_reg=AdaBoostClassifier()
ada_reg.fit(X_train,Y_train)
```

```python
bugg_reg=,BaggingClassifier()
bugg_reg.fit(X_train,Y_train)
```

```python
print("RandomForestClassifier score",reg.score(X_test,Y_test))
print("GradientBoostingClassifier score",grad_reg.score(X_test,Y_test))
print("AdsBoostClassifier",ada_reg.score(X_test,Y_test))
print("BuggingClassifier",bugg_reg.score(X_test,Y_test))
```

```python
y_pred1=reg.predict(X_test)
y_pred2=grad_reg.predict(X_test)
y_pred3=ada_reg.predict(X_test)
y_pred4=bugg_reg.predict(X_test)
```

```python
cm1 = confusion_matrix(Y_test, y_pred1)
cm2= confusion_matrix(Y_test, y_pred2)
cm3 = confusion_matrix(Y_test, y_pred3)
cm4 = confusion_matrix(Y_test, y_pred4)
print("confusion_matrix of RandomForestClassifier model",cm1)
print("confusion_matrix of GradientBoostingClassifier model",cm2)
print("confusion_matrix of AdaBoostClassifier model",cm3)
print("confusion_matrix of BuggingClassifier model",cm4)
```
