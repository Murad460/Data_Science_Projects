```python
df_train=pd.read_csv("/kaggle/input/chronic-kidney-disease/kidney_disease_train.csv")
```

```python
df_train=df_train.fillna('o')
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df_train[["rbc","pc","pcc","ba","htn","dm","cad","appet","pe","ane"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
```

```python
concaty=pd.concat([df_train,one],axis=1).drop(columns=["rbc","pc","pcc","ba","htn","dm","cad","appet","pe","ane","classification"])
label_encoder=LabelEncoder()
# Y=df_train[["classification"]
Y=label_encoder.fit_transform(df_train["classification"])
```

```python
# Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(one,Y,test_size=0.2,random_state=42)
# X_test,Y_test=train_test_split(one_test,Y_test,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```
