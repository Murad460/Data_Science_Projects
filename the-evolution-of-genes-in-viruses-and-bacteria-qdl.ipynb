```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,BaggingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import HuberRegressor,RANSACRegressor,BayesianRidge
from sklearn.svm import SVR
```

```python
df=pd.read_csv("/kaggle/input/the-evolution-of-genes-in-viruses-and-bacteria/NCBI_Escherichia_coli_genomes.csv")
df
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=["GenBank FTP","RefSeq FTP","Replicons","Release Date"])
df=df.dropna(how="any",axis=0)
df
```

```python
dd=df["Organism Groups"].str.split(";",expand=True)
dd
```

```python
concaty=pd.concat([df,dd],axis=1).drop(columns=["Organism Groups"])
concaty
```

```python
concaty["Strain"]
```

```python
label_encoder=LabelEncoder()
for col in concaty.columns:
    concaty[col]=label_encoder.fit_transform(concaty[col])
```

```python
concaty["Bacteria"]=concaty[0]
concaty["proteoBacteria"]=concaty[1]
concaty["Gammaproteobacteria"]=concaty[2]
concaty=concaty.drop(columns=[0,1,2])
```

```python
concaty
```

```python
X=concaty.drop(columns=["Strain"])
Y=concaty["Strain"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
for i in [RandomForestRegressor(),GradientBoostingRegressor(),AdaBoostRegressor(),BaggingRegressor(),BayesianRidge(),KNeighborsRegressor(),DecisionTreeRegressor(),SVR(),HuberRegressor(),RANSACRegressor()]:
    random_reg=i
    random_reg.fit(X_train,Y_train)
    print(f"the accuracy of {i}",random_reg.score(X_test,Y_test))
```

# ##Deep Learning  

```python
from keras.layers import Layer,Dense,Flatten,BatchNormalization,Input,LeakyReLU
from keras.initializers import HeNormal,GlorotUniform
from keras.regularizers import l2,l1_l2
from keras.callbacks import EarlyStopping
from keras.models import Model,Sequential
from keras.optimizers import Adam
```

```python
inputs=Input(shape=(13,))
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
d=Dense(64,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dense(128,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dense(228,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dense(528,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dense(1028,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
outputs=Dense(1)(d)
model=Model(inputs,outputs)
optimizer=Adam(learning_rate=0.001,beta_1=0.99,beta_2=0.99)
model.compile(optimizer=optimizer,loss="mse",metrics=["mae"])
model.fit(X_train,Y_train,epochs=15,validation_data=(X_test,Y_test))
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
```

```python
import cirq
import tensorflow as tf
```

```python
def quantum_circuit(data):
    qubit=[cirq.GridQubit(0,i) for i in range(13)]
    circuit=cirq.Circuit()
    for i,value in enumerate(data):
        q=i%len(qubit)
        circuit.append(cirq.H(qubit[q]))
        circuit.append(cirq.rx(value*np.pi).on(qubit[q]))
        circuit.append(cirq.ry(value*np.pi).on(qubit[q]))
        if (i+1)%len(qubit)<len(qubit):
            circuit.append(cirq.CNOT(qubit[q],qubit[(i+1)%len(qubit)]))
            circuit.append(cirq.measure(qubit[q],key=f"qubit_{i}"))
    return circuit,qubit
def run_quantum_circuit(data):
    results=[]
    simulator=cirq.Simulator()
    for inputs_data in data:
        circuit,qubit=quantum_circuit(inputs_data)
        result=simulator.run(circuit)
        measure0=[]
        for i in range(len(qubit)):
            measure0.append(result.measurements[f"qubit_{i}"].flatten()[0])
        results.append(measure0)
    return np.array(results,dtype=np.float32)
class QuantumLayer(Layer):
    def __init__(self):
        super(QuantumLayer,self).__init__()
    def call(self,inputs):
        outputs=tf.numpy_function(run_quantum_circuit,[inputs],tf.float32)
        outputs.set_shape(inputs.shape)
        return outputs
model2=Sequential([
    QuantumLayer(),
    Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    Dense(64,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dense(128,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dense(228,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dense(528,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dense(1028,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dense(1)
])
model2.compile(optimizer="adam",loss="mae",metrics=["mae"])
model2.fit(X_train,Y_train,epochs=15,validation_data=(X_test,Y_test),batch_size=32)
```
