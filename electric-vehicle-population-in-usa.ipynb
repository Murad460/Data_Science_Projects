# # AutoEncoder

```python
from keras.layers import Dense,LSTM,GRU,Flatten,Input,BatchNormalization,ReLU,Dropout
from keras.models import Model
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
new_df_reshaped=df.iloc[0:20000,:].values
new_df_reshaped=np.repeat(new_df_reshaped[:,np.newaxis,:],20,axis=1)
new_df_reshaped.shape
```

```python
inputs=Input(shape=(20,13491))
L=LSTM(32,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=GRU(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=GRU(256,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=GRU(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=LSTM(32,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
outputs=LSTM(13491,activation="softmax",return_sequences=True)(L)
model=Model(inputs,outputs)
model.summary()
```

```python
 model.fit(new_df_reshaped,new_df_reshaped,epochs=10,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```
