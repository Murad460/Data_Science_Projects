```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier,GradientBoostingRegressor
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
cols=['Course_Name', 'Category', 'Duration (hours)',
       'Enrolled_Students', 'Completion_Rate (%)', 'Platform', 'Price ($)',
       'Rating (out of 5)']
df=pd.read_csv("/kaggle/input/online-courses-usage-and-history-dataset/online_courses_uses.csv",usecols=cols)
```

```python
label_encod=LabelEncoder()
lab=["Course_Name","Category","Platform"]
for l in lab:
    df[l]=label_encod.fit_transform(df[l])
```

```python
one_hot=OneHotEncoder(sparse_output=False)
dy=one_hot.fit_transform(df[["Course_Name","Category","Platform"]])
dy=pd.DataFrame(dy,columns=one_hot.get_feature_names_out())
```

```python
dy
```

```python
X=df.drop(columns=["Rating (out of 5)"])
Y=df["Rating (out of 5)"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_reg=RandomForestRegressor(n_estimators=50)
random_reg.fit(X_train,Y_train)
random_reg.score(X_test,Y_test)
```

```python
grad_reg=GradientBoostingRegressor()
grad_reg.fit(X_train,Y_train)
grad_reg.score(X_test,Y_test)
```
