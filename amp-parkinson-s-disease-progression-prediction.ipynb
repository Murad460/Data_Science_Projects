# # Deep learning

```python
from keras.layers import Dense,BatchNormalization,Dropout,Input
from keras.models import Model
from keras.regularizers import l2
from keras.initializers import HeNormal
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor="val_loss",
patience=5,
verbose=1)
```

```python
X_train.shape
```

```python
inputs=Input(shape=(7,))
D=Dense(32,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(inputs)
D=Dense(32,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)
B=BatchNormalization()(D)

D=Dense(32,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(B)
D=Dense(32,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)

B=BatchNormalization()(D)
D=Dropout(0.5)(B)
D=Dense(64,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)
D=Dense(64,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)
B=BatchNormalization()(D)

D=Dense(64,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(B)
D=Dense(64,activation='relu',kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)


outputs=Dense(3,activation="softmax")(D)
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss="sparse_categorical_crossentropy",metrics=["accuracy"])

```

```python
model.fit(X_train,Y_train,epochs=50,batch_size=32,validation_data=(X_test,Y_test),callbacks=[early_stopping])
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
```
