```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/kaggle/input/credit-card-transactions-dataset/credit_card_transactions.csv",nrows=100000)
df=df.drop(columns=["merch_zipcode"])
```

```python
df["dob"]=pd.to_datetime(df["dob"])
df["Year"]=df["dob"].dt.year
df["Month"]=df["dob"].dt.month
df["Day"]=df["dob"].dt.day
```

```python
df["trans_date_trans_time"]=pd.to_datetime(df["trans_date_trans_time"])
df["Year_trans"]=df["trans_date_trans_time"].dt.year
df["Month_trans"]=df["trans_date_trans_time"].dt.month
df["Day_trans"]=df["trans_date_trans_time"].dt.day
df["Hour_trans"]=df["trans_date_trans_time"].dt.hour
df["Minutes_trans"]=df["trans_date_trans_time"].dt.minute
df["Seconds_trans"]=df["trans_date_trans_time"].dt.second
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["gender","merchant","category","first","last","street","city","state","job"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
```

```python
concaty=pd.concat([df,one],axis=1).drop(columns=["Unnamed: 0","gender","trans_date_trans_time","merchant","cc_num","is_fraud","category","first","dob","last","street","city","state","job","trans_num"])
```

```python
concaty
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,df["is_fraud"],test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```
