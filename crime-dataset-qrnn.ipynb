# # QRNN

```python
from keras.layers import Layer,Input,Dropout,LSTM,BatchNormalization,TimeDistributed,ReLU,LeakyReLU,Dense
from keras.models import Model,Sequential
from keras.initializers import HeNormal
from keras.regularizers import l2
import cirq
import tensorflow as tf
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
def quantum_circuit(data):
    qubit=[cirq.GridQubit(0,i) for i in range(3)]
    circuit=cirq.Circuit()
    for i,value in enumerate(data):
#         value=value*2
        circuit.append(cirq.rx(value*2*np.pi).on(qubit[i%len(qubit)]))
        circuit.append(cirq.ry(value*2*np.pi).on(qubit[i%len(qubit)]))
        circuit.append(cirq.rz(value*2*np.pi).on(qubit[i%len(qubit)]))
    circuit.append([cirq.CNOT(qubit[i % len(qubit)], qubit[(i + 1) % len(qubit)]) for i in range(len(data) - 1)])
    circuit.append([cirq.measure(qubit[i%len(qubit)],key=f"qubit_{i}") for i in range(len(data))])
    return circuit,qubit
def run_quantum_circuit(data):
    results=[]
    for input_data in data:
        circuit,qubit=quantum_circuit(input_data)
        simulator=cirq.Simulator()
        result=simulator.run(circuit)
        measure0=[]
        for i in range(len(input_data)):
            measure0.append(result.measurements[f'qubit_{i}'].flatten()[0])
        results.append(measure0)
    return np.array(results,dtype=np.float32)
class QuantumLayer(Layer):
    def __init__(self):
        super(QuantumLayer,self).__init__()
    def call(self,inputs):
#         input_flat=tf.reshape(inputs,[-1,inputs.shape[-1]])
        output=tf.numpy_function(run_quantum_circuit,[inputs],tf.float32)
        output.set_shape((inputs.shape[0],inputs.shape[1]))  # Get the batch size at runtime
        return output

model=Sequential([
    QuantumLayer(),
    Dense(32,activation="relu"),
    Dense(1,activation="sigmoid")
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
model.fit(X_train,Y_train,epochs=5,batch_size=32)
```
