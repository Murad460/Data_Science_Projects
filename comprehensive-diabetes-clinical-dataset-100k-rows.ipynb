```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/kaggle/input/100000-diabetes-clinical-dataset/diabetes_dataset.csv")
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["gender","location","smoking_history"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
# one
```

```python
concaty=pd.concat([df,one],axis=1).drop(columns=["gender","location","smoking_history","diabetes"])
```

```python
concaty
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,df["diabetes"],test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```
