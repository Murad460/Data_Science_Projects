```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df_train=pd.read_csv("/kaggle/input/playground-series-s4e3/train.csv")
df_test=pd.read_csv("/kaggle/input/playground-series-s4e3/test.csv")
```

```python
df_train
```

```python
X_train_data=df_train.drop(columns=["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"])
Y_train_data=df_train[["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"]]
```

```python
X_train,X_test=train_test_split(X_train_data,test_size=0.2,random_state=42)
Y_train,Y_test=train_test_split(Y_train_data,test_size=0.2,random_state=42)
```

# # Deep learninig

```python
from keras.layers import Dense,Dropout,BatchNormalization,Input,SimpleRNN,LSTM,GRU,Flatten
from keras.models import Model
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
X_train.shape
```

# # M1

```python
inputs=Input(shape=(28,))
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
d=Dropout(0.2)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
b=BatchNormalization()(d)
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(b)
d=Dropout(0.2)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
outputs=Dense(7,activation="softmax")(d)
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,Y_train,epochs=50,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

```python
for layer in model.layers:
    layer.trainable=False
model_output=model.layers[-1].output
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
```

# # M2

```python
inputs=Input(shape=(28,))
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(model_output)
d=Dropout(0.2)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
b=BatchNormalization()(d)
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(b)
d=Dropout(0.2)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)

d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=Dropout(0.2)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
b=BatchNormalization()(d)
d=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(b)
d=Dropout(0.2)(d)
d=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)

outputs2=Dense(7,activation="softmax")(d)
model2=Model(inputs=model.input,outputs=outputs2)
model2.compile(optimizer='adam',loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model2.fit(X_train,Y_train,epochs=50,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

```python
loss,accuracy=model2.evaluate(X_test,Y_test)
```

```python
for layer2 in model2.layers:
    layer2.trainable=False
model2_output=model2.layers[-1].output
```

# # M3

```python
inputs=Input(shape=(28,))
d=Dense(320,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(model2_output)
d=Dropout(0.2)(d)
d=Dense(640,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
b=BatchNormalization()(d)
d=Dense(320,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(b)
d=Dropout(0.2)(d)
d=Dense(640,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)

d=Dense(320,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=Dropout(0.2)(d)
d=Dense(640,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
b=BatchNormalization()(d)
d=Dense(320,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(b)
d=Dropout(0.2)(d)
d=Dense(640,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)

outputs3=Dense(7,activation="softmax")(d)
model3=Model(inputs=model.input,outputs=outputs3)
model3.compile(optimizer='adam',loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model3.fit(X_train,Y_train,epochs=50,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

```python
loss,accuracy=model3.evaluate(X_test,Y_test)
```

# # # M4 time series model

```python
X_train_reshaped=X_train.values.reshape(X_train.shape[0],1,X_train.shape[1])
X_test_reshaped=X_test.values.reshape(X_test.shape[0],1,X_test.shape[1])
X_train_reshaped.shape
```

```python
inputs4=Input(shape=(1,28))
S=SimpleRNN(32,activation="relu",return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs4)
S=SimpleRNN(32,activation="relu",return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(S)
b=BatchNormalization()(S)
d=Dropout(0.2)(b)

S=SimpleRNN(32,activation="relu",return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
S=SimpleRNN(32,activation="relu",return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(S)
b=BatchNormalization()(S)
d=Dropout(0.2)(b)
f=Flatten()(d)
d=Dense(32,activation="relu")(f)
d=Dense(10,activation='relu')(d)
outputs4=Dense(7,activation="softmax")(d)
model4=Model(inputs4,outputs4)
model4.compile(optimizer='adam',loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model4.fit(X_train_reshaped,Y_train,epochs=50,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

```python
loss,accuracy=model4.evaluate(X_test_reshaped,Y_test)
```
