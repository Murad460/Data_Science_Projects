```python
d=df.dropna(how='any',axis=1)
d
```

```python
new_df=pd.get_dummies(d).astype(int)
```

```python
new_df_reshape=new_df.values
new_df_reshape = np.reshape(new_df_reshape, (new_df_reshape.shape[0], 1, new_df_reshape.shape[1]))
new_df_reshape.shape
```

# # AutoEncoder

```python
from keras.layers import Dense,LSTM,GRU,Flatten,Reshape,BatchNormalization,Input,ReLU,Dropout
from keras.models import Model
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
inputs=Input(shape=(1,18948))
L=LSTM(32,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)
L=LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)
L=LSTM(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)
L=LSTM(256,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)

L=LSTM(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)
L=LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)
L=LSTM(32,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=BatchNormalization()(L)
L=Dropout(0.5)(L)
outputs = LSTM(18948, return_sequences=True, activation="sigmoid")(L)  
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"])
model.summary()
```

```python
model.fit(new_df_reshape,new_df_reshape,epochs=5,batch_size=32,validation_split=0.2)
```

```python
from sklearn.metrics import *
predict_normal=model.predict(new_df_reshape)
reconstruction_error_normal=np.mean(np.square(new_df_reshape-predict_normal))
threshold=np.mean(reconstruction_error_normal)+2*np.std(reconstruction_error_normal)

new_df_reshape_test=new_df_reshape[100:1200]
predict_test=model.predict(new_df_reshape_test)
reconstruction_error_test=np.mean(np.square(new_df_reshape_test-predict_test))

label=np.where(reconstruction_error_test>threshold,0,1)
new_df_reshape_test=new_df_reshape_test.flatten()
predict_test=label.flatten()
accuracy_score(new_df_reshape_test,predict_test)
```
