```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingRegressor
```

```python
df=pd.read_csv("/kaggle/input/geolifeclef-2022-lifeclef-2022-fgvc9/observations/observations_fr_train.csv",delimiter=";")
df=df.drop(columns=["subset"])
```

```python
df.isnull().sum()
```

```python
Y=df["species_id"]
X=df.drop(columns=["species_id"])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_reg=RandomForestClassifier()
random_reg.fit(X_train,Y_train)
random_reg.score(X_test,Y_test)
```

```python
grad_reg=GradientBoostingRegressor()
grad_reg.fit(X_train,Y_train)
grad_reg.score(X_test,Y_test)
```
