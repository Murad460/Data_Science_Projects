```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import *
```

```python
df=pd.read_csv("/kaggle/input/iriscsv/Iris.csv")
```

```python
df.isnull().sum()
```

```python
new_df=pd.get_dummies(df).astype(int)
```

```python
new_df
```

```python
X=new_df.drop(columns=["Species_Iris-setosa","Species_Iris-versicolor","Species_Iris-virginica"])
Y=new_df[["Species_Iris-setosa","Species_Iris-versicolor","Species_Iris-virginica"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
for i in [RandomForestClassifier()]:
    classif=i
    classif.fit(X_train,Y_train)
    classif.score(X_test,Y_test)
    y_pred=classif.predict(X_test)
    print(f" accuracy of {i}",accuracy_score(Y_test,y_pred))
```

```python
for i in [RandomForestClassifier(),GradientBoostingClassifier(),AdaBoostClassifier(),SVC()]:
    multioutput=MultiOutputClassifier(i)
    multioutput.fit(X_train,Y_train)
    multioutput.score(X_test,Y_test)
    y_pred=multioutput.predict(X_test)
    print(f"accuracy of {i}:",accuracy_score(Y_test,y_pred))
```
