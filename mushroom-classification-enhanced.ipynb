```python
import  numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/kaggle/input/mushroom-classification-enhanced/mushroom.csv")
```

```python
df.columns
```

```python
columns=['cap-shape', 'cap-surface', 'cap-color', 'ruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']
```

```python
transformer_numeric=Pipeline(steps=[
    ("simple_imputer",SimpleImputer(strategy="constant",fill_value="missing"))
    
])
transformer_onehot=Pipeline(steps=[
        ("One_hot",OneHotEncoder())])
```

```python
columns_transformer=ColumnTransformer([
    ("transformer_numeric",transformer_numeric,columns),
    ("transformer_onehot",transformer_onehot,columns)
])
```

```python
pipeline=Pipeline(steps=[
    ("preprocessing",columns_transformer)
])
dd=pipeline.fit_transform(df)
```

```python
categorical_transformer = Pipeline(steps=[
    ("simple_imputer", SimpleImputer(strategy="constant", fill_value='missing')),
    ("one_hot", OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[

        ("cat", categorical_transformer, columns)
    ]
)
```

```python
pipeline=Pipeline(steps=[
    ("preprocessing",preprocessor)
])
dd=pipeline.fit_transform(df)
dd=dd.toarray()
dd=pd.DataFrame(dd,columns=pipeline.get_feature_names_out())
```

```python
dd["class"]=df["class"]
label=LabelEncoder()
dd["class"]=label.fit_transform(dd["class"])
dd
```

```python
X=dd.drop(columns=["class"])
Y=dd[["class"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier(n_estimators=50)
random_classif.fit(X_train,Y_train)
```

```python
random_classif.score(X_test,Y_test)
```
