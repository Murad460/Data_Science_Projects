```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download("punkt")
nltk.download("stopwords")
```

```python
df=pd.read_csv("/kaggle/input/fashion-nova-reviews/fashionnova_reviews.csv",nrows=50000)
df=df.drop(columns=["Reviewer Name","Profile Link","Review Date","Date of Experience"])
```

```python
df=df.fillna("nil")
df
```

```python
def text_removing(text):

    data_re=re.sub(r'\D+',' ',text)
    return data_re
def text_preprocessing(text):
    data=text.lower()
    data_re=re.sub(r'\W+',' ',data)
    word_data_re=word_tokenize(data_re)
    tokens=[word for word in word_data_re if word not in stopwords.words("english")]
    return "  ".join(tokens)
```

```python
df["Review Count"]=df["Review Count"].apply(text_removing)
df["Rating"]=df["Rating"].apply(text_removing)
```

```python
df["Review Title"]=df["Review Title"].apply(text_preprocessing)
```

```python
df["Review Text"]=df["Review Text"].apply(text_preprocessing)
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["Country"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
```

```python
tfid=TfidfVectorizer()
rtext=tfid.fit_transform(df["Review Text"])
rtext=rtext.toarray()
rtext=pd.DataFrame(rtext,columns=tfid.get_feature_names_out())
# rtitle=tfid.fit_transform(df[["Review Title"]])
# rtitle=pd.DataFrame(rtitle,columns=tfid.get_feature_names_out())
tfid2=TfidfVectorizer()
rtext2=tfid2.fit_transform(df["Review Title"])
rtext2=rtext2.toarray()
rtext2=pd.DataFrame(rtext2,columns=tfid2.get_feature_names_out())
```

```python
rtext
```

```python
concaty=pd.concat([one,rtext,rtext2],axis=1)
```

```python
concaty.columns=[i for i,_ in enumerate(concaty.columns)]
```

```python
concaty
```

```python
Y=df["Review Count"]
# X=Concaty
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
grad_classif=GradientBoostingClassifier()
grad_classif.fit(X_train,Y_train)
grad_classif.score(X_test,Y_test)
```
