```python
train_label=pd.read_csv("/kaggle/input/ariel-data-challenge-2024/train_labels.csv")
train_adc=pd.read_csv("/kaggle/input/ariel-data-challenge-2024/train_adc_info.csv")
```

```python
label=train_label.drop(columns=["planet_id"])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(train_adc,label,test_size=0.2,random_state=42)
```

# # Deep learning

```python
from keras.layers import Dense,LSTM,SimpleRNN,GRU,Dropout,Input,BatchNormalization
from keras.models import Model
from keras.regularizers import l2
from keras.initializers import HeNormal
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
X_train.shape
X_train_reshaped=X_train.values.reshape(X_train.shape[0],1,X_train.shape[1])
X_test_reshaped=X_test.values.reshape(X_test.shape[0],1,X_test.shape[1])
X_train_reshaped.shape
```

```python
Y_train
```

```python
Y_train.nunique().sum()
```

```python
inputs=Input(shape=(1,6))
S=SimpleRNN(32,return_sequences=True,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
S=SimpleRNN(32,return_sequences=False,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(S)
D=Dense(64,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(S)
D=Dense(32,activation="relu",kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(D)
outputs=Dense(283)(D)
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss='mse',metrics=["mae"])
```

```python
model.fit(X_train_reshaped,Y_train,epochs=50,batch_size=32,validation_data=(X_test_reshaped,Y_test),callbacks=[early_stopping])
```

```python
loss,mae=model.evaluate(X_test_reshaped,Y_test)
```
