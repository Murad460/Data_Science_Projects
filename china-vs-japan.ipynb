```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/spotify-song-performance-dataset/spotify_data.csv')
```

```python
df.isnull().sum()
```

```python
 df=df.fillna("0")
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df).astype(int)
```

```python
df
```

# # AutoEncoder

```python
from keras.layers import Dense,Flatten,LSTM,Dropout,SimpleRNN,GRU,Input,BatchNormalization,ReLU
from keras.models import Model
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
train_df=df.values
train_df=train_df.reshape(train_df.shape[0],1,train_df.shape[1])
train_df.shape
```

```python
inputs=Input(shape=(1,4970))
L=LSTM(32,activation="relu",return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(inputs)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=LSTM(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)

L=LSTM(256,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)

L=LSTM(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)

L=LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
L=LSTM(32,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(L)
L=ReLU()(L)
L=Dropout(0.5)(L)
L=BatchNormalization()(L)
outputs=LSTM(1,activation="sigmoid",return_sequences=True)(L)
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"])
model.summary()
```

```python
model.fit(train_df,train_df,epochs=10,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

```python
from sklearn.metrics import *
test_data=train_df[100:200]
prediction_test=model.predict(test_data)
   
binary_prediction=(prediction_test>0.5).astype(int)
prediction=model.predict(train_df)
reconstruction_normal=np.mean(np.square(train_df-prediction),axis=1)
reconstruction_test=np.mean(np.square(test_data-prediction_test),axis=1)

threshold=np.mean(reconstruction_normal)+2*np.std(reconstruction_normal)
label=np.where(reconstruction_test>threshold,0,1)
print(label)
test_data=test_data.flatten()
binary_prediction=binary_prediction.flatten()


print(accuracy_score(test_data,binary_prediction))
```
