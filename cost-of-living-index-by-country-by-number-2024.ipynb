```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/cost-of-living-index-by-country-by-number-2024/Cost_of_Living_Index_by_Country_2024.csv")
```

```python
df
```

```python
X=df.drop(columns=["Country"])
Y=df[["Country"]]
```

```python
df.plot()
```

```python
plt.figure(figsize=(8,7),)
df.hist()
```

```python
plt.figure(figsize=(11,7))
df.boxplot()
```

```python

df_melted = pd.melt(df, id_vars=Y, value_vars=X, var_name='Variable', value_name='Value')
plt.figure(figsize=(12,7))
# Create a violin plot
sns.violinplot(x='Variable', y='Value', data=df_melted, palette='muted', inner='quartile')

# Show the plot
plt.show()
```

```python

df_melted = pd.melt(df, id_vars=Y, value_vars=X, var_name='Variable', value_name='Value')
plt.figure(figsize=(12,7))
# Create a violin plot
sns.scatterplot(x='Variable', y='Value', data=df_melted, palette='muted')

# Show the plot
plt.show()
```

```python
df_pie=pd.melt(df,id_vars=Y,value_vars=X,var_name="variable",value_name="value")
plt.pie(X["Cost of Living Index"],labels=Y,autopct='%1.1f%%', colors=['skyblue', 'orange', 'lightgreen'])
```

```python

# Create pie charts for each value column
columns =X.columns 

for col in columns:
    plt.figure(figsize=(8, 8))
    plt.pie(df[col], labels=Y, autopct='%1.1f%%', colors=['skyblue', 'orange', 'lightgreen'])
    plt.title(f'Pie Chart for {col}')
    plt.show()
```
