```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/kaggle/input/mountains-vs-beaches-preference/mountains_vs_beaches_preferences.csv")
```

```python
df=pd.get_dummies(df).astype(int)
```

```python
df
```

```python
X=df.drop(columns=["Preference"])
Y=df["Preference"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```
