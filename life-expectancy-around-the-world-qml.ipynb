```python
dp=pd.read_csv("/kaggle/input/life-expectancy-around-the-world/probability-that-a-newborn-will-die-in-the-first-year-of-life-by-sex.csv")
```

```python
dp.isnull().sum()
```

```python
dp["Code"]=dp["Code"].fillna("nil")
```

```python
new_dp=pd.get_dummies(dp).astype(int)
```

```python
new_dp
```

```python
X=new_dp.drop(columns=["Probability of death - Type: period - Sex: female - Age: 0"])
Y=new_dp["Probability of death - Type: period - Sex: female - Age: 0"]
```

```python
Y.nunique()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```
