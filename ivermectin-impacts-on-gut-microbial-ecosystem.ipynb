```python
new_df2=df2.drop(columns=["sampleID","study day","sample_type"])
```

```python
one_hot=OneHotEncoder()
d=one_hot.fit_transform(new_df2)
dg=d.toarray()
pf=pd.DataFrame(dg,columns=one_hot.get_feature_names_out())
pf["study_day"]=df2["study day"]
label_encoder=LabelEncoder()
l=label_encoder.fit_transform(df2["sample_type"])
# ld=l.toarray()
pf["sample_ty"]=pd.DataFrame(l)
pf
```

```python
random_classifier=RandomForestClassifier(n_estimators=50)
random_classifier.fit(X_train,Y_train)
```

```python
random_classifier.score(X_test,Y_test)
```

```python
import cirq
import tensorflow as tf
import tensorflow_quantum as tfq
import sympy 
import numpy
from keras.layers import Dense,Flatten,Input
from keras.models import Model
from tensorflow.dtypes import string
from tensorflow_quantum.layers import PQC, convert_to_tensor
```

```python
X_train.shape[1]
```

```python
def convert_to_circuit(X_train):
    qubits=[cirq.GridQubit(0,i) for i in range(X_train.shape[0])]
    circuits=[]
    for x in X_train:
        circuit=cirq.Circuit()
        for i,value in enumerate(x):
            if value:
                circuit.append(cirq.X(qubits[i]))
            circuits.append(circuit)
        return qubits,circuits
qubits,circuits=convert_to_circuit(X_train)
def create_model(qubits):
    symbols=sympy.symbols(f'theta0:{len(qubits)}')
    for i,qubit in enumerate(qubits):
        circuit=cirq.Circuit()
        circuit.append(cirq.rx(symbols[i])(qubit))
    readout_operators=[cirq.Z(qubit) for qubit in qubits]
    inputs=Input(shape=(),dtype=string)
    pqc_output = tfq.layers.PQC(model_circuit=circuit, operators=readout_operators)
    outputs = pqc_output(inputs)

    outputs = Dense(1)(outputs)
    model=Model(inputs=inputs,outputs=outputs)
    return model
model1=create_model(qubits)
```
