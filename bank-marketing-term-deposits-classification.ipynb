```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/kaggle/input/bank-marketing-term-deposits-classification/Train.csv")
```

```python
df
```

```python
df.columns
```

```python
data=['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'poutcome']
```

```python
transformer=ColumnTransformer(transformers=[
    ("one_hot",OneHotEncoder(),data)
])
td=transformer.fit_transform(df)
td1=td.toarray()
td2=pd.DataFrame(td1,columns=transformer.get_feature_names_out())
new_df1=pd.concat([df,td2],axis=1)
new_df=new_df1.drop(columns=['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'poutcome'])
```

```python
new_df
```

```python
df.dtypes
```

```python
X=new_df.drop(columns=["y"])
Y=df[["y"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
label_encoder=LabelEncoder()
new_Y1=label_encoder.fit_transform(Y_train)
new_Y=pd.DataFrame(new_Y1,columns=["target"])
# new_Y
```

```python
random_class=RandomForestClassifier(n_estimators=50)
random_class.fit(X_train,Y_train)
random_class.score(X_test,Y_test)
```

# # Deep learning

```python
from keras.layers import Dense,Flatten,Dropout,LSTM,SimpleRNN,Input
from keras.models import Model
```

```python
X_train_reshaped=X_train.values.reshape(X_train.shape[0],1,X_train.shape[1])
X_train_reshaped.shape
```

```python
inputs=Input(shape=(1,51))
l1=LSTM(32,activation="relu",return_sequences=True)(inputs)
l2=LSTM(64,activation="relu")(l1)
d2=Dense(128,activation="relu")(l2)
outputs=Dense(1,activation="sigmoid")(d2)
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train_reshaped,new_Y,epochs=10,batch_size=42,validation_split=0.2)
```
