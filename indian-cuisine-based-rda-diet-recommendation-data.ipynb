# # Deep learning model LSTM

```python
X_train_reshaped=X_train.values.reshape(X_train.shape[0],1,X_train.shape[1])
X_train_reshaped.shape
```

```python
X_train.dtypes
```

```python
Y_train = np.array(Y_train, dtype=np.float32)
Y_train.shape
```

```python
from keras.layers import Dense,Input,Flatten,Dropout,LSTM
from keras.models import Model,Sequential
```

```python
inputs=Input(shape=(1,14))
l1=LSTM(32,activation="relu")(inputs)
l2=Dense(32,activation="relu")(l1)
outputs=Dense(len(Y_train),activation="softmax")(l2)
model=Model(inputs,outputs)
model.compile(optimizer="adam",loss="mse",metrics=["accuracy"])
```

```python
model.fit(X_train_reshaped,Y_train,epochs=20,batch_size=32)
```
