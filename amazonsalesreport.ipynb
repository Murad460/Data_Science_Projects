```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/kaggle/input/amazonsalesreport/Amazon Sale Report.csv")
```

```python
df.columns
```

```python
len(df)
```

```python
df.sample()
```

```python
df.isnull().sum()
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["Status","Fulfilment","Sales Channel","ship-service-level","Category","Size","Courier Status","currency","ship-city","ship-state","B2B"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
# one["Amount"]=df["Amount"]
# one["ship-postal-code"]=df["ship-postal-code"]
one["Qty"]=df["Qty"]
```

```python
X=one.drop(columns=["Status_Shipped - Rejected by Buyer"])
Y=one[["Status_Shipped - Rejected by Buyer"]]
```

```python
Y.isnull().sum().sum()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
len(X_test)
```

```python
X1=X_train.iloc[0:25000]
Y1=Y_train.iloc[0:25000]

X2=X_train.iloc[25000:50000]
Y2=Y_train.iloc[25000:50000]

X3=X_train.iloc[50000:75000]
Y3=Y_train.iloc[50000:75000]

X4=X_train.iloc[75000:100000]
Y4=Y_train.iloc[75000:100000]

X5=X_train.iloc[100000:128000]
Y5=Y_train.iloc[100000:128000]
```

```python
sgd_class=SGDClassifier()
sgd_class.partial_fit(X1,Y1,classes=[0,1])
```

```python
sgd_class.partial_fit(X2,Y2)
sgd_class.score(X_test.iloc[0:25000],Y_test.iloc[0:25000])
```

```python
sgd_class.partial_fit(X3,Y3)
sgd_class.score(X_test.iloc[0:25000],Y_test.iloc[0:25000])
```

```python
sgd_class.partial_fit(X4,Y4)
sgd_class.score(X_test.iloc[0:25000],Y_test.iloc[0:25000])
```

```python
sgd_class.partial_fit(X5,Y5)
sgd_class.score(X_test.iloc[:25000],Y_test.iloc[0:25000])
```

```python
y_pred=sgd_class.predict(X_test)
```
