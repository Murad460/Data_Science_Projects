```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("/kaggle/input/weight-training/strong.csv")
```

```python
df.nunique()
```

```python
df["Seconds"].unique()
```

```python
df.isnull().sum()
```

```python
df['RPE'].fillna(df['RPE'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
new_df=df.drop(columns=["Date","Notes","Workout Notes"])
```

# # one hot ecoder

```python
one_hot=OneHotEncoder(sparse=False)
one=one_hot.fit_transform(new_df[["Workout Name","Exercise Name"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
```

```python
new_df["Duration"]=new_df["Duration"].dt.total_seconds()
new_df
```

```python
concaty=pd.concat([new_df,one],axis=1)
```

```python
X=concaty.drop(columns=["RPE","Workout Name","Exercise Name"])
Y=concaty[["RPE"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
new_df.dtypes
new_df
```

```python
random_reg=RandomForestRegressor()
random_reg.fit(X_train,Y_train)
random_reg.score(X_test,Y_test)
```
