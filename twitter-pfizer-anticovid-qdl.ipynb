```python
import numpy as np
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from keras.utils import to_categorical
nltk.download("punkt")
nltk.download("stopwords")
```

```python
df=pd.read_csv("/kaggle/input/twitter-pfizer-anticovid/dataset_Vaccine_Pfizer.csv")
```

```python
new_df=df.dropna(how='any',axis=0)
new_df.isnull().sum()
```

```python
new_df
```

```python
def text_preprocess(text):
    data=text.lower()
    data_re=re.sub(r'\W+',' ',data)
    data_word_tokenize=word_tokenize(data_re)
    tokens=[word for word in data_word_tokenize if word not in stopwords.words("english")]
    return " ".join(tokens)

def text_vectorizer(text):
    vectorizer=TfidfVectorizer()
    vect=vectorizer.fit_transform(text)
    
    vect_feature=pd.DataFrame(vect.toarray(),columns=vectorizer.get_feature_names_out())
    return vect_feature

```

```python
new_df["Text"]=new_df["Text"].apply(text_preprocess)
dd=text_vectorizer(new_df["Text"])
```

```python
dd
```

```python
concaty=pd.concat([df,dd],axis=1).drop(columns=["Text","Target"])
concaty=concaty.dropna(how='any',axis=0)
```

```python
label_encoder=LabelEncoder()
label=label_encoder.fit_transform(new_df["Target"])
label=to_categorical(label)
label.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,label,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
import tensorflow as tf

import cirq
from keras.layers import Layer,Dense,Flatten
from keras.models import Model,Sequential
from keras.callbacks import EarlyStopping
```

```python
def build_quantum_circuit(data):
    qubit=[cirq.GridQubit(0,i) for i in range(len(data))]
    circuit=cirq.Circuit()
    for i,value in enumerate(data):
        circuit.append(cirq.rx(value*np.pi).on(qubit[i]))
        circuit.append(cirq.ry(value*np.pi).on(qubit[i]))
        circuit.append(cirq.rz(value*np.pi).on(qubit[i]))
    circuit.append(cirq.CNOT(qubit[0],qubit[1]))
    circuit.append([cirq.measure(qubit[i],key=f'qubit_{i}')for i in range(len(data))])
    return circuit,qubit
def run_quantum_circuit(data):
    results=[]
    for inputs_data in data:
        circuit,qubit=build_quantum_circuit(inputs_data)
        simulator=cirq.Simulator()
        result=simulator.run(circuit)
        measure0=result.measurements["qubit_0"].flatten()[0]
        measure1=result.measurements["qubit_1"].flatten()[0]
        combine_result=np.array([measure0,measure1])
        results.append(combine_result)
    return np.array(results,np.float32)
class QuantumLayer(Layer):
    def __init__(self):
        super(QuantumLayer,self).__init__()
    def call(self,inputs):
        output=tf.numpy_function(run_quantum_circuit,[inputs],tf.float32)
        output.set_shape((None,2))
        return output
model=Sequential([
    QuantumLayer(),
    Dense(3,activation="sigmoid")
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train,Y_train)
```
