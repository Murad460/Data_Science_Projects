```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
```

```python
df=pd.read_csv("/kaggle/input/playground-series-s4e4/train.csv")
```

```python
df
```

```python
label_encoder=LabelEncoder()
df["Sex2"]=label_encoder.fit_transform(df["Sex"])
```

```python
X=df.drop(columns=["Sex","Rings"])
Y=df["Rings"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
grad_classif=GradientBoostingClassifier()
grad_classif.fit(X_train,Y_train)
grad_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```

```python
deci_classif=DecisionTreeClassifier()
deci_classif.fit(X_train,Y_train)
deci_classif.score(X_test,Y_test)
```

```python
svm_classif=SVC()
svm_classif.fit(X_train,Y_train)
svm_classif.score(X_test,Y_test)
```

```python
kn_classif=KNeighborsClassifier()
kn_classif.fit(X_train,Y_train)
kn_classif.score(X_test,Y_test)
```

```python
bernoulli_classif=BernoulliNB()
bernoulli_classif.fit(X_train,Y_train)
bernoulli_classif.score(X_test,Y_test)
```

# # Deep learning

```python
from keras.layers import Dense,Input,BatchNormalization,Dropout
from keras.models import Model
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
Y_train.nunique()
```

```python
print(max(Y_train))
print(max(Y_test))
```

```python
X_train.shape
```

```python
inputs=Input(shape=(9,))
D=Dense(32,activation="relu",kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(inputs)
D=Dense(32,activation="relu",kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)
B=BatchNormalization()(D)
D=Dropout(0.5)(B)
D=Dense(32,activation="relu",kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)
D=Dense(32,activation="relu",kernel_regularizer=l2(0.01),kernel_initializer=HeNormal())(D)
outputs=Dense(30,activation="softmax")(D)
model=Model(inputs,outputs)
model.compile(optimizer='adam',loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,Y_train,epochs=50,batch_size=32,validation_data=(X_test,Y_test),callbacks=[early_stopping])
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
```
