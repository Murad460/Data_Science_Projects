```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,BaggingRegressor
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
```

```python
df="/kaggle/input/nasa-nearest-earth-objects-1910-2024/nearest-earth-objects(1910-2024).csv"
```

```python
pd_df=pd.read_csv("/kaggle/input/nasa-nearest-earth-objects-1910-2024/nearest-earth-objects(1910-2024).csv")
```

```python
pd_df.isnull().sum()
```

```python
simple_imp=SimpleImputer(strategy="mean")
d=simple_imp.fit_transform(pd_df[["absolute_magnitude","estimated_diameter_min","estimated_diameter_max"]])
ob=pd.DataFrame(d,columns=pd_df[["absolute_magnitude","estimated_diameter_min","estimated_diameter_max"]].columns)
ob.isnull().sum()
oz=pd_df.drop(columns=["absolute_magnitude","estimated_diameter_min","estimated_diameter_max"])
```

```python
new_df=pd.concat([oz,ob],axis=1)
new_df
```

```python
lab_encoder=LabelEncoder()
new_df["lab_orbiting_body"]=lab_encoder.fit_transform(new_df["orbiting_body"])
new_df["is_hazardous_n"]=lab_encoder.fit_transform(new_df["is_hazardous"])
new_df1=new_df.drop(columns=["orbiting_body","is_hazardous"])
new_df1
X=new_df1.drop(columns=["name","is_hazardous_n"])
Y=new_df1[["is_hazardous_n"]]
new_df1
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
X_train.shape
```

```python
random_reg=RandomForestRegressor()
random_reg.fit(X_train,Y_train)
```

```python
random_reg.score(X_test,Y_test)
```

```python
deci_reg=DecisionTreeRegressor()
deci_reg.fit(X_train,Y_train)
deci_reg.score(X_test,Y_test)
```

```python
grad_reg=GradientBoostingRegressor()
grad_reg.fit(X_train,Y_train)
grad_reg.score(X_test,Y_test)
```

```python
ada_reg=AdaBoostRegressor()
ada_reg.fit(X_train,Y_train)
ada_reg.score(X_test,Y_test)
```

```python
bag_reg=BaggingRegressor()
bag_reg.fit(X_train,Y_train)
bag_reg.score(X_test,Y_test)
```
