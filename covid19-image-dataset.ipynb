```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df="/kaggle/input/covid19-image-dataset/Covid19-dataset/train"
```

```python
new_df=[]
for cur,dirs,filename in os.walk(df):
    for file in filename:
        data=os.path.join(cur,file)
        cv=cv2.imread(data)
        cv_resize=cv2.resize(cv,(250,250))
        if cv_resize is not None:
            new_df.append(cv_resize)
        
de=np.array(new_df)
```

```python
de.dtype
```

```python
de.shape
```

```python
y=[]
for _ in range(111):
    i=1
    y.append(i)
len(y)
```

```python
for _ in range(70):
    o=0
    y.append(o)
len(y)
```

```python
for _ in range(70):
    k=2
    y.append(k)
len(y)
    
```

```python
X=de
y1=np.array(y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y1,test_size=0.2,random_state=42)
```

```python
random_class=RandomForestClassifier(n_estimators=50)
random_class.fit(X_train,Y_train)
random_class.score(X_test,Y_test)
```

# # Deep learning

```python
X_train.shape
```

```python
from keras.layers import Dense,Flatten,Dropout,Input,Conv2D,MaxPooling2D,UpSampling2D
from keras.models import Model,Sequential
```

```python
model = Sequential([
    Input(shape=(250,250,3)),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(256, (3, 3), activation='relu', padding='same'),
    UpSampling2D(size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    UpSampling2D(size=(2, 2)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Flatten(),
    Dense(64,activation="relu"),
    Dense(3,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
Y_train_reshaped = np.expand_dims(Y_train, axis=-1)
Y_train_reshaped.shape
```

```python
model.fit(X_train,Y_train)
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
```
