```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/kaggle/input/playground-series-s3e20/train.csv")
df=df.drop(columns=["ID_LAT_LON_YEAR_WEEK"])
```

```python
df.dtypes
```

```python
df.tail(10)
```

```python
impute=SimpleImputer(strategy="mean")
impute_df=impute.fit_transform(df)
impute_df=pd.DataFrame(impute_df,columns=impute.get_feature_names_out())
```

```python
impute_df
```

```python
df.iloc[:,0:20].isnull().sum()
```

```python
X=impute_df.drop(columns=["emission"])
Y=impute_df["emission"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
random_reg=RandomForestRegressor()
random_reg.fit(X_train,Y_train)
random_reg.score(X_test,Y_test)
```

```python
linear_reg=LinearRegression()
linear_reg.fit(X_train,Y_train)
linear_reg.score(X_test,Y_test)
```

```python
grad_reg=GradientBoostingRegressor()
grad_reg.fit(X_train,Y_train)
grad_reg.score(X_test,Y_test)
```
