```python
lab_encod=LabelEncoder()
onehot_encod=OneHotEncoder()
d=onehot_encod.fit_transform(df[["data"]])

d1=onehot_encod.fit_transform(df[["route_type"]])
d2=onehot_encod.fit_transform(df[["is_cutoff"]])
d3=onehot_encod.fit_transform(df[["trip_uuid"]])
d4=onehot_encod.fit_transform(df[["route_schedule_uuid"]])
```

```python
# d1=d.toarray
df["data"]=pd.DataFrame(d,columns=["data"])
df["route_type"]=pd.DataFrame(d1,columns=[["route_type"]])
df["is_cutoff"]=pd.DataFrame(d2,columns=[["is_cutoff"]])
df["trip_uuid"]=pd.DataFrame(d3,columns=[["trip_uuid"]])
df["route_schedule_uuid"]=pd.DataFrame(d3,columns=[["route_schedule_uuid"]])
```
