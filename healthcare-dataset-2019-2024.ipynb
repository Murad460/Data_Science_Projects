```python
from keras.layers import Dense,Dropout,LSTM,Input
from keras.models import Model
from keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(
monitor='val_loss',
patience=5,
verbose=1)
```

```python
inputs=Input(shape=(39907,))
d1=Dense(32,activation="relu")(inputs)
d1=Dense(64,activation="relu")(d1)
d=Dropout(0.5)(d1)
d1=Dense(128,activation="relu")(d)
d1=Dense(64,activation="relu")(d1)
outputs=Dense(1,activation="sigmoid")(d1)
model=Model(inputs,outputs)
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(Xt,Yt,batch_size=32,epochs=100,validation_split=0.2,callbacks=[early_stopping])
model.fit(Xt2,Yt2,batch_size=32,epochs=100,validation_split=0.2,callbacks=[early_stopping])
model.fit(Xt3,Yt3,batch_size=32,epochs=100,validation_split=0.2,callbacks=[early_stopping])
model.fit(Xt4,Yt4,batch_size=32,epochs=100,validation_split=0.2,callbacks=[early_stopping])
```
