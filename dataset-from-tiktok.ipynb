```python
col=["claim_status","video_transcription_text","verified_status","author_ban_status"]
def fill_nil(df,col):
    for cl in col:
        df[cl]=df[cl].replace(0,"nil")
    return df
dj=fill_nil(df,col)
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(dj[["claim_status","author_ban_status"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
one
```

# # text preprocess

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download("stopwords")
nltk.download("punkt")
```

```python
def vectorizer_data(df,col):
    vectorizer=TfidfVectorizer()
    vectorizer_df=[]
    for cl in col:
        transformer=vectorizer.fit_transform(df[cl])
        transformer=transformer.toarray()
        vectorizer_df.append(pd.DataFrame(transformer,columns=vectorizer.get_feature_names_out()))
    final_data=pd.concat(vectorizer_df,axis=1)
    return final_data
```

```python
col2=["video_transcription_text"]

dt=vectorizer_data(df,col2)
```

```python
dt
```

```python
concaty=pd.concat([dj,one,dt],axis=1).drop(columns=["claim_status","video_transcription_text","verified_status","author_ban_status"])
label_encoder=LabelEncoder()
Y=label_encoder.fit_transform(df["verified_status"])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(concaty,Y,test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
grad_classif=GradientBoostingClassifier()
grad_classif.fit(X_train,Y_train)
grad_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```

```python
svm_classif=SVC()
svm_classif.fit(X_train,Y_train)
svm_classif.score(X_test,Y_test)
```
