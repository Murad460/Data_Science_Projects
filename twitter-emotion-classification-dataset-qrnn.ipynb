```python
import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from nltk.corpus import stopwords
nltk.download("punkt")
nltk.download("stopwords")
```

```python
df=pd.read_parquet("/kaggle/input/twitter-emotion-classification-dataset/train-00000-of-00001.parquet")
df=df.sample(n=15000)
```

```python
df
```

```python
def text_preprocess(text):
    data=text.lower()
    data_re=re.sub(r"\W+"," ",data)
    data_word=word_tokenize(data_re)
    tokens=[word for word in data_word if word not in stopwords.words("english")]
    return " ".join(tokens)
def text_vectorizer(text):
    vectorizer=CountVectorizer()
    data=text.apply(text_preprocess)
    da=vectorizer.fit_transform(data)
    da=pd.DataFrame(da.toarray(),columns=vectorizer.get_feature_names_out())
    return da
```

```python
da=text_vectorizer(df["text"])
```

```python
da.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(da,df["label"],test_size=0.2,random_state=42)
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```

```python
log_classif=LogisticRegression()
log_classif.fit(X_train,Y_train)
log_classif.score(X_test,Y_test)
```

# # QDL

```python
import cirq
import tensorflow as tf
from keras.layers import Layer,Dense,BatchNormalization,Input,LeakyReLU,Dropout,SimpleRNN,GRU,LSTM,Flatten
from keras.initializers import HeNormal
from keras.regularizers import l2
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam
```

```python
early_stopping=EarlyStopping(
monitor="val_loss",
verbose=1,
patience=5,
restore_best_weights=True)
```

```python
X_train.shape
```

```python
X_train_reshape=X_train.values
X_train_reshape=X_train_reshape.reshape(X_train.shape[0],1,X_train.shape[1])
X_test_reshape=X_test.values
X_test_reshape=X_test_reshape.reshape(X_test.shape[0],1,X_test.shape[1])
X_test_reshape.shape
```

```python
def quantum_circuit(data):
    qubit=[cirq.GridQubit(0,i) for i in range(10)]
    circuit=cirq.Circuit()
    for i,value in enumerate(data):
        q=i%len(qubit)
        circuit.append(cirq.H(qubit[q]))
        circuit.append(cirq.rx(value*2*np.pi).on(qubit[q]))
        circuit.append(cirq.X(qubit[q]))
        circuit.append(cirq.ry(value*2*np.pi).on(qubit[q]))
        circuit.append(cirq.Y(qubit[q]))
        circuit.append(cirq.rz(value*2*np.pi).on(qubit[q]))
        if (i+2)%len(qubit)<len(qubit):
            circuit.append(cirq.CCX(qubit[q],qubit[(i+1)%len(qubit)],qubit[(i+2)%len(qubit)]))
            circuit.append(cirq.CNOT(qubit[q],qubit[(i+2)%len(qubit)]))
        circuit.append(cirq.measure(qubit[q],key=f"qubit_{i}"))
    return circuit,qubit
def run_quantum_circuit(data):
    results=[]
    for inputs_data in data:
        circuit,qubit=quantum_circuit(inputs_data)
        simulator=cirq.Simulator()
        result=simulator.run(circuit)
        measure0=[]
        for i in range(len(inputs_data)):
            measure0.append(result.measurements[f"qubit_{i}"].flatten()[0])
        results.append(measure0)
    return np.array(results,dtype=np.float32)
class QuantumLayer(Layer):
    def __init__(self):
        super(QuantumLayer,self).__init__()
    def call(self,inputs):
        output=tf.numpy_function(run_quantum_circuit,[inputs],tf.float32)
        output.set_shape(inputs.shape)
        return output
model=Sequential([
    QuantumLayer(),
    Dense(32,activation='relu'),
    Dense(64,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    Dense(128,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    Dense(256,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    BatchNormalization(),
    Dense(556,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    BatchNormalization(),
    Dense(1056,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    BatchNormalization(),
#     Dense(128,activation='relu'),

    Dense(1,activation="sigmoid")
    
])
optimizer=Adam(learning_rate=0.001,beta_1=0.9,beta_2=0.9)
model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=["accuracy"])
```

```python
model.fit(X_train.iloc[:400,:6],Y_train[:400],epochs=10,batch_size=32,validation_data=(X_test.iloc[:400,:6],Y_test[:400]),callbacks=[early_stopping])
```

```python
class QuantumLayer2(Layer):
    def __init__(self):
        super(QuantumLayer2,self).__init__()
    def call(self,inputs):
        batch_size = tf.shape(inputs)[0]
        feature_dim = tf.shape(inputs)[2]
        
        inputs_flat = tf.reshape(inputs, [batch_size, feature_dim])
#         inputs_flat=tf.reshape(inputs,[inputs.shape[0],1*inputs.shape[2]])
        output=tf.numpy_function(run_quantum_circuit,[inputs_flat],tf.float32)
        output = tf.reshape(output, [batch_size, 1, feature_dim])
        
        output.set_shape([None, 1, feature_dim])
#         output.set_shape(inputs.shape)
        return output
model2=Sequential([
    QuantumLayer2(),
    SimpleRNN(32,activation='relu',return_sequences=True),
    
    LSTM(64,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    
    LSTM(128,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    
    GRU(256,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    
    GRU(556,return_sequences=True,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    
    GRU(1056,return_sequences=False,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01)),
    LeakyReLU(0.08),
    Dropout(0.5),
    BatchNormalization(),
    Flatten(),
#     Dense(128,activation='relu'),

    Dense(1,activation="sigmoid")
    
])
optimizer=Adam(learning_rate=0.001,beta_1=0.9,beta_2=0.9)
model2.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=["accuracy"])
model2.fit(X_train_reshape[:4,:4],Y_train[:4],epochs=10,batch_size=32,validation_data=(X_test_reshape[:4,:4],Y_test[:4]),callbacks=[early_stopping])
```
