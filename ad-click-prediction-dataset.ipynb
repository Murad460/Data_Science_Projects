```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/kaggle/input/ad-click-prediction-dataset/ad_click_dataset.csv")
df=df.fillna("nil")
df["age"]=df["age"].replace("nil",0)
```

```python
one_hot=OneHotEncoder(sparse_output=False)
one=one_hot.fit_transform(df[["full_name","gender","device_type","ad_position","browsing_history","time_of_day"]])
one=pd.DataFrame(one,columns=one_hot.get_feature_names_out())
one["id"]=df["id"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(one,df["click"],test_size=0.2,random_state=42)
```

```python
df["click"].nunique()
```

```python
random_classif=RandomForestClassifier()
random_classif.fit(X_train,Y_train)
random_classif.score(X_test,Y_test)
```
