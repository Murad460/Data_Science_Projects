```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
df=pd.read_csv("/kaggle/input/specific-language-impairment/all_data_R_high_cor_removed.csv")
```

```python
df.isnull().sum()
```

```python
df=df.dropna(how="any",axis=0)
df.isnull().sum()
```

```python
col=df.columns
label_encoder=LabelEncoder()
for c in col:
    if df[c].dtype=="object":
        df[c]=label_encoder.fit_transform(df[c])
df
```

```python
df.nunique()
```

```python
X=df.drop(columns=["total_error"])
Y=df["total_error"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
Y_train
```

```python
for i in [RandomForestRegressor(),GradientBoostingRegressor(),KNeighborsRegressor()]:
    random_reg=i
    random_reg.fit(X_train,Y_train)
    print(f"the accuracy of {i}",random_reg.score(X_test,Y_test))
```

# # deep learning# 

```python
import tensorflow as tf
from keras.layers import Layer,Dense,Flatten,BatchNormalization,Dropout,Input,LeakyReLU
from keras.models import Model,Sequential
from keras.initializers import HeNormal,GlorotUniform
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
```

```python
inputs=Input(shape=(49,))
d=Dense(32,activation="relu")(inputs)
d=Dense(64,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dropout(0.5)(d)
d=Dense(128,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dense(256,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dropout(0.5)(d)
d=BatchNormalization()(d)
d=Dense(32,activation="relu")(d)
d=Dense(64,kernel_initializer=GlorotUniform(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dropout(0.5)(d)
d=Dense(128,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dense(256,kernel_initializer=HeNormal(),kernel_regularizer=l2(0.01))(d)
d=LeakyReLU(0.08)(d)
d=Dropout(0.5)(d)
d=BatchNormalization()(d)
outputs=Dense(1)(d)
model=Model(inputs,outputs)
model.compile(optimizer="adam",loss="mse",metrics=["mae"])
model.summary()
```

```python
model.fit(X_train,Y_train,epochs=15,batch_size=32,validation_data=(X_test,Y_test))
```

```python
import cirq
```

```python
def quantum_circuit(data):
    qubit=[cirq.GridQubit(0,i) for i in range(5)]
    circuit=cirq.Circuit()
    for i,value in enumerate(data):
        q=i%len(qubit)
        circuit.append(cirq.rx(value*np.pi).on(qubit[q]))
        circuit.append(cirq.H(qubit[q]))
        circuit.append(cirq.ry(value*np.pi).on(qubit[q]))
        if (i+1)%len(qubit)<len(qubit):
            circuit.append(cirq.CNOT(qubit[q],qubit[(i+1)%len(qubit)]))
            circuit.append(cirq.measure(qubit[q],key=f"qubit_{i}"))
    return circuit,qubit
def run_quantum_circuit(data):
    results=[]
    simulator=cirq.Simulator()
    for inputs_data in data:
        circuit,qubit=quantum_circuit(inputs_data)
        result=simulator.run(circuit)
        measure0=[]
        for i in range(len(inputs_data)):
            measure0.append(result.measurements[f"qubit_{i}"].flatten()[0])
        results.append(measure0)
    return np.array(results,dtype=np.float32)
class QuantumLayer(Layer):
    def __init__(self):
        super(QuantumLayer,self).__init__()
    def call(self,inputs):
        outputs=tf.numpy_function(run_quantum_circuit,[inputs],tf.float32)
        outputs.set_shape(inputs.shape)
        return outputs
model=Sequential([
    QuantumLayer(),
    Dense(1)
    
])
model.compile(optimizer="adam",loss="mse",metrics=["mae"])
model.fit(X_train.iloc[:49,:49],Y_train[:49],epochs=15,batch_size=32)
```
