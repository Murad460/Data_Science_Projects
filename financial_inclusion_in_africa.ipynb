```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/content/Train.csv")
```

```python
df["country"].unique()
```

```python
df.nunique()
```

```python
X=df.drop(columns=["uniqueid","bank_account"]).sample(frac=1,random_state=42)
Y=df["bank_account"].sample(frac=1,random_state=42)
```

```python
X["country"].unique()
```

```python
train_encoders = [] # Dictionary to store LabelEncoders for each column
for col in X.columns:
    encoder = LabelEncoder()
    X[col] = encoder.fit_transform(X[col])  # Transform each column in-place
    train_encoders.append(encoder)
```

```python
train_encoders
```

```python
label_encoder=LabelEncoder()
Y=label_encoder.fit_transform(Y)
```

```python
Y
```

```python
random_forest=RandomForestClassifier()
random_forest.fit(X,Y)
random_forest.score(X,Y)
```
